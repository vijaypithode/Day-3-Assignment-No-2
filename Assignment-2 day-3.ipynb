# 1. ifelse and elif used to write program python for report cards 

```python
 sub1=int(input("Enter marks subject 1:-"))
 sub2=int(input("Enter marks subject 2:-"))
 sub3=int(input("Enter marks subject 3:-"))
 sub4=int(input("Enter marks subject 4:-"))
 sub5=int(input("Enter marks subject 5:-"))
avg=sub1+sub2+sub3+sub4+sub5/5
if avg>=90:
    print("A Grade")
elif avg>=70 and avg<=90:
    print("B Grade")
elif avg>=60 and avg<=70:
    print("C Grade")
elif avg>=35 and avg<=60:
    print("D Grade")
else:
        print("Failed")
```

# 2. use for loop to print prime number

```python
for i in range(1,1000):
    
    for j in range(2,i):
        if i % j == 0:
             break;
        
    else: 
        print(i)
```
